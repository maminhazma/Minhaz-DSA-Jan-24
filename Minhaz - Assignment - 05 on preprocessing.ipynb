#4 Fill in missing values

```python
#To fill the missing values in column Age, Cabin, Embarked
```

```python
mean_age = data1['Age'].mean() # age missing values replace with mean of the column
data1['Age'].fillna(mean_age, inplace=True)
```

```python
mode_cabin = data1['Cabin'].mode() # cabin missing value replace with mode value
print (mode_cabin)
```

```python
mode_cabin_fill = mode_cabin.iloc[2] # cabin missing value replace with 3rd mode value
data1['Cabin'].fillna(mode_cabin_fill, inplace=True)
```

```python
mode_embarked = data1['Embarked'].mode() # embarked missing value replace withe mode value
print (mode_embarked)
```

```python
mode_embarked_fill = mode_embarked[0] # embarked missing value replace with first mode value
data1['Embarked'].fillna(mode_embarked_fill, inplace=True)
```

```python
data1.isnull().sum() # all missing values cleared 
```
